# アンサンブル

`バギング`：同じデータと同じモデルを複数用意しても意味がない、違うデータと違うモデル同士で比較を行う必要があるよね。計算方法はn個のモデルの予測値の平均。  
`ブートストラップ`：訓練データからランダムに（重複あり）n個のデータをサンプリングすること  
`スタッキング`：バギングはただの平均であり、個々の予測値の重要度を考慮できていない、、、スタッキングは重要度も考慮する。（重みづけ）  
`バンピング`：複数のモデルの中から最も当てはまりの良いモデルを探す。いっぱいモデルを作って一番よいモデルを採用する。一見メリットなさそうだけど、質の悪いデータを用いて望ましくない解が得られてしまったとき、それらのデータを除いたブートストラップデータ集合によって、よりよい解が得られることがある。  
`ランダムフォレスト`：バギングベースの学習器して「決定木」を用いた手法。１．訓練データからn個のブートストラップデータ集合をとる。２．これらを使って訓練データの組み合わせを一部選ぶ。（全部の組み合わせデータは使わない。）３．回帰の場合は平均を、分類の場合は多数決をとって最終的な予測値とする。  

#### なぜベース学習器に決定木を用いるのか？
バギングの基本的な考え方は、分散大・バイアス小のモデルを複数組み合わせて誤差を減らすこと。  
　　分散大・バイアス小　→　複雑なモデル（決定木、最近傍法など）  
　　分散小・バイアス大　→　単純なモデル（線形回帰）  
決定木はバギングのベース学習器としては理想的な分散大・バイアス小のモデル。  
その他、「高速である」、「変数のデータ型を問わない」、「スケーリングに対して不変」などのメリットがある。  

#### なぜ一部の特徴量だけしか使わないのか
モデル間の相関が低ければ低いほど最終的な予測値の制度は高まる。異なったデータで学習したモデルを組み合わあわせたほうが性能がたかい、ブートストラップに加え、それぞれのモデルで学習に使う変数を変えることでモデル間の相関を低くする。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
moons = make_moons(n_samples=200, noise=0.2, random_state=0)
x = moons[0]
y = moons[1]
```

```python
from matplotlib.colors import ListedColormap

def plot_decision_boundary(model, x, y, margin=0.3):
    _x1 = np.linspace(x[:,0].min()-margin, x[:,0].max()+margin, 100)
    _x2 = np.linspace(x[:,1].min()-margin, x[:,1].max()+margin, 100)
    x1, x2 = np.meshgrid(_x1, _x2)
    x_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = model.predict(x_new).reshape(x1.shape)
    custom_cmap = ListedColormap(['mediumblue', 'orangered'])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap)
    
def plot_datasets(x, y):
    plt.plot(x[:,0][y==0], x[:,1][y==0], 'bo', ms=15)
    plt.plot(x[:,0][y==1], x[:,1][y==1], 'r^', ms=15)
    plt.xlabel('$x_0$', fontsize=30)
    plt.ylabel('$x_1$', fontsize=30, rotation=0)

plt.figure(figsize=(12, 8))
plot_datasets(x, y)
plt.show()
```

### 決定木でまずやってみる、これは直線で分類されるため滑らかではないのが弱点

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

tree_clf = DecisionTreeClassifier().fit(x_train, y_train)
```

```python
plt.figure(figsize=(12, 8))
plot_decision_boundary(tree_clf, x, y)
plot_datasets(x, y)
plt.show()
```

### random-forestで滑らかに表現できる

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(n_estimators=100, random_state=0).fit(x_train, y_train)

plt.figure(figsize=(12, 8))
plot_decision_boundary(random_forest, x, y)
plot_datasets(x, y)
plt.show()
```

### 違うデータでやってみる

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
x_iris = iris.data
y_iris = iris.target
```

```python
plt.figure(figsize=(12, 8))
plt.barh(range(iris.data.shape[1]), random_forest_iris.feature_importances_, height=0.5)
plt.yticks(range(iris.data.shape[1]), iris.feature_names, fontsize=20)
plt.xlabel('Feature importances', fontsize=30)
plt.show()
```

### 違うデータでやってみる　タイタニックデータセット

```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.head()
```

#### ageとembarkedの欠損値補完　　cabinは少なすぎるため今回は外す

```python
df.info()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())  # 平均値で保管
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])  # 最頻値で保管
```

```python
df.info()
```

### ダミー変数　　sexとembarkedをダミー変数にする

```python
from sklearn.preprocessing import LabelEncoder

cat_features = ['Sex', 'Embarked']

for col in cat_features:
    lbl = LabelEncoder()
    df[col] = lbl.fit_transform(list(df[col].values))
```

```python
df.head()
```

### 訓練データとテストデータ

```python
x = df.drop(columns=['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'])
y = df['Survived']
```

```python
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
tree = DecisionTreeClassifier().fit(x_train, y_train)
tree.score(x_test, y_test)
```

```python
rnd_forest = RandomForestClassifier(n_estimators=500, max_depth=5, random_state=0).fit(x_train, y_train)
```

```python
rnd_forest.score(x_test, y_test)
```
